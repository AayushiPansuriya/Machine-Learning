```python
#box blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("img1.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#apply box blur
box=cv2.blur(img,(10,10))
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(box)
```

```python
#Gaussian blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("img3.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#apply gaussian blur
box=cv2.GaussianBlur(img,(9,9),0)
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(box)
```

```python
#median blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("img4.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#apply median blur
med=cv2.medianBlur(img,(11))
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(med)
```

```python
#bluring an image using inbuilt functions
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("img4.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#apply box blur
box=cv2.blur(img,(10,10)) # we can have any positive number

#apply gaussian blur
gaus=cv2.GaussianBlur(img,(15,15),5) # it can be any od number,
                                      #0 meansbautomatically calculate the blur effect(you can have any number instead of 0)

#apply median blur
med=cv2.medianBlur(img,(11))

plt.subplot(2,2,1),plt.imshow(img),plt.title('source image')
plt.subplot(2,2,2),plt.imshow(box),plt.title('box blur image')
plt.subplot(2,2,3),plt.imshow(gaus),plt.title('gauss blur image')
plt.subplot(2,2,4),plt.imshow(med),plt.title('med blur image')
```

```python
#motion blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("img4.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

size=20

#(i).horizontal blur
k1=np.zeros((size,size))
k1[int(size/2),:]=np.ones(size)
k1=k1/size

#(i).vertical blur
k2=np.zeros((size,size))
k2[:,int(size/2)]=np.ones(size)
k2=k2/size

#applying the kernel
horizontal=cv2.filter2D(img,-1,k1)
vertical=cv2.filter2D(img,-1,k2)

#display the img
plt.subplot(1,3,1),plt.imshow(img),plt.title('source image')
plt.subplot(1,3,2),plt.imshow(horizontal),plt.title('horizontal blur')
plt.subplot(1,3,3),plt.imshow(vertical),plt.title('vertical blur')
```
