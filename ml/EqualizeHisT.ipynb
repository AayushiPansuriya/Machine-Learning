```python
#constrasting the gray scale image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('flower.jpg',0)
img1=cv2.equalizeHist(img)

plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.title('source img'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(img1,cmap='gray'),plt.title('equalizehist on gray'),plt.axis('off')
```

```python
#constrasting the color image
#using global histogram equalization
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('flower1.jpg')

#convert an image to YCrCb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

#split the color channel
y,cr,cb=cv2.split(ycrcb)

#global histogram equalization
y_eq=cv2.equalizeHist(y)
ycrcb_eq=cv2.merge([y_eq,cr,cb])

#convert an image to BGR
global_eq=cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)
#convert an image from BGR TO RGB
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
global_eq1=cv2.cvtColor(global_eq,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('source img'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_eq1),plt.title('equalizehist on color'),plt.axis('off')
```

```python
#constrasting the color image
#using clahe
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('flower1.jpg')

#convert an image to YCrCb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

#split the color channel
y,cr,cb=cv2.split(ycrcb)

#clahe
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe=clahe.apply(y)
ycrcb_clahe=cv2.merge([y_clahe,cr,cb])

#convert an image to BGR
clahe_eq=cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)
#convert an image from BGR TO RGB
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
clahe_eq1=cv2.cvtColor(clahe_eq,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('source img'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(clahe_eq1),plt.title('clahe on color'),plt.axis('off')
```
