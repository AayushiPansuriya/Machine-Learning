```python
#Erosion and Dilation
import cv2
import numpy as np
import matplotlib.pyplot as plt

#load image
img=cv2.imread('image1.png')

k1=np.ones((3,3),np.uint8)

#Erosian
erosian=cv2.erode(img,k1,iterations=1)

#Dilation
dilation=cv2.dilate(img,k1,iterations=1)

plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.title('source')
plt.subplot(1,3,2),plt.imshow(erosian,cmap='gray'),plt.title('erosian')
plt.subplot(1,3,3),plt.imshow(dilation,cmap='gray'),plt.title('dilation')
```
